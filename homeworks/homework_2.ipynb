## Задание 1

Напишите код на Python в среде Jupyter Notebook для решения следующей задачи.

Дана переменная, в которой хранится слово из латинских букв. Напишите код, который выводит на экран:

* среднюю букву, если число букв в слове нечётное;
* две средних буквы, если число букв чётное.


Пример работы программы:

* 
    ```python
    word = 'test’
    ```
    Результат:
    ```python
    es
    ```
*  
    ```python
    word = 'testing’
    ```  
    Результат:
    ```python
    t
    ```

````python
def solution(word: str) -> str:
    word_len = len(word)
    return word[
        (
            word_len // 2 - 1 if word_len % 2 == 0
            else word_len // 2
        ):word_len // 2 + 1
    ]
````

````python
testing_function = solution

test_cases = [
    ({'word': 'test'}, 'es'),
    ({'word': 'testing'}, 't'),
    ({'word': 'work'}, 'or'),
    ({'word': 'Shrek'}, 'r')
]


for test_data, expected_result in test_cases:
    print(
        f'Тестовые данные:\n{test_data}\n',
        f'Ожидаемый результат:\n{expected_result}\n',
        f'Результат теста: {testing_function(**test_data) == expected_result}',
        sep='\n', end='\n\n\n'
    )
````

## Задание 2 (необязательное)

Напишите код на Python в среде Jupyter Notebook для решения следующей задачи.

Вы делаете MVP (минимально жизнеспособный продукт) dating-сервиса.
У вас есть список юношей и девушек.
Выдвигаем гипотезу: лучшие рекомендации получатся, если просто отсортировать имена по алфавиту и познакомить людей с одинаковыми индексами после сортировки. Но вы не будете никого знакомить, если кто-то может остаться без пары.

Примеры работы программы:

* 
    ```python
    boys = ['Peter', 'Alex', 'John', 'Arthur', 'Richard']
    girls = ['Kate', 'Liza', 'Kira', 'Emma', 'Trisha']
    ```

    Результат
    ```python
    Идеальные пары:
    Alex и Emma
    Arthur и Kate
    John и Kira
    Peter и Liza
    Richard и Trisha
    ```

* 
    ```python
    boys = ['Peter', 'Alex', 'John', 'Arthur', 'Richard', 'Michael']
    girls = ['Kate', 'Liza', 'Kira', 'Emma', 'Trisha']
    ```

    Результат: 
    ```python
    Внимание, кто-то может остаться без пары.
    ```

````python
def solution(boys: list, girls: list) -> str:
    if len(boys) == len(girls):
        return '\n'.join(
            ['Идеальные пары:']
            + [
                f'{girl} и {boy}'
                for girl, boy in zip(sorted(boys), sorted(girls))
            ]
        )
    else:
        return 'Внимание, кто-то может остаться без пары.'
````

````python
testing_function = solution

test_cases = [
    (
        {
            'boys': ['Peter', 'Alex', 'John', 'Arthur', 'Richard'],
            'girls': ['Kate', 'Liza', 'Kira', 'Emma', 'Trisha']
        },
        (
            'Идеальные пары:\n'
            'Alex и Emma\n'
            'Arthur и Kate\n'
            'John и Kira\n'
            'Peter и Liza\n'
            'Richard и Trisha'
        )
    ),
    (
        {
            'boys': ['Peter', 'Alex', 'John', 'Arthur', 'Richard', 'Michael'],
            'girls': ['Kate', 'Liza', 'Kira', 'Emma', 'Trisha']
        },
        'Внимание, кто-то может остаться без пары.'
    )
]


for test_data, expected_result in test_cases:
    print(
        f'Тестовые данные:\n{test_data}\n',
        f'Ожидаемый результат:\n{expected_result}\n',
        f'Результат теста: {testing_function(**test_data) == expected_result}',
        sep='\n', end='\n\n\n'
    )
````
